```python
# Import Dependencies
import requests
import pandas as pd
```

```python
# Instantiate Function
def getInfo(year):
    
    # Base URL
    url = f'http://ice.nhl.com/ice/playerfantasystats.htm?season={year}&gameType=2&team=&position=L&viewName=gameMinimums'
    
    # Establish Header
    header = {
        "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
        "X-Requested-With": "XMLHttpRequest"
    }
    
    # Get Request
    r = requests.get(url, headers=header)
    
    # Read HTML
    df = pd.read_html(r.text)
    
    # Create DataFrame
    new_df = pd.DataFrame(df[2])
    
    # Return DataFrame
    return new_df
```
